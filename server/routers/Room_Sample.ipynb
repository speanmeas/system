```python
import random
import string


def generate_room_data(num_rooms):
    statuses = ["Available", "Booked", "Dirty"]
    room_types = ["Standard", "Deluxe", "Suite"]
    prices = [random.uniform(50, 200) for _ in range(num_rooms)]
    capacities = [random.randint(1, 5) for _ in range(num_rooms)]
    room_nos = [f"{i:04}" for i in range(1, num_rooms+1)]

    return [
        {
            "room_number": room_no,
            "room_type": random.choice(room_types),
            "status": random.choice(statuses),
            "price": price,
            "capacity": capacity,
        }
        for room_no, price, capacity in zip(room_nos, prices, capacities)
    ]


sample_data = generate_room_data(10000)


# pprint(sample_data)
```
